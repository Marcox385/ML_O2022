## 4.2 Tuplas

Son como listas, pero sus elementos son fijos, una vez que la tupla es creada, no es possible añadir nuevos elementos, borrar elementos, reemplazar elementos, o reordenar los elementos en la tupla.

Una razón por la que existen las tuplas es por su desempeño. 
El procesamiento de una tupla es más rápido que el de una lista, entonces las tuplas son buena elección cuando procesas lotes de información y la información no será modificada.



## 4.3 Diccionarios

- Un diccionario es un objeto que almacena un conjunto de información. 
- Cada elemento en un diccionario tiene dos partes: una clave y un valor. 
- Se utiliza una clave para localizar un valor específico.
- Para acceder a un valor de un diccionario se utiliza la clave con el siguiente formato:
        
        Dictionary_name[key]

### Algunos métodos de diccionarios

- $clear$: borra el contenido de un diccionario
- $get$: obtiene el valor asociado con un $key$ específico. Si el $key$ no es encontrado, el método devuelve un valor por default.
- $items$:devuelve todoas las claves ($keys$) de un diccionario y sus valores asociados como una secuencia de tuplas.
- $keys$: devuelve todas las claves ($keys$) de un diccionario como una secuencia de tuplas
- $pop$: devuelve el valor asociado con una clave específica y borra el par clave-valor del diccionario. Si la clave no se encuentra regresa un valor por default.
- $popitem$: devuelve un par de clave-valor aleatoriamente seleccionado como una tupla y borra el elemento.
- $values$: devuelve todos los valore en el diccionario como una secuencia de tuplas.
