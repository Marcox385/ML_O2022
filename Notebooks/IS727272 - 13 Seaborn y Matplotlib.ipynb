## 13.4 Regresiones Lineales

En estadística el análisis de regresión lineal es un proceso estadístico para estimar la relación entre variables, es decir, cómo varía el valor de una variable en función del valor de otras variables.

```python
# Leer una base de datos de propinas que está en nuestra librería Seaborn

propinas = sns.load_dataset('tips')
propinas.head(10)
```

```python
propinas.info()
```

```python
sns.lmplot('total_bill', 'tip', propinas)
```

```python
propinas['porcentaje'] = 100 * (propinas['tip'] / propinas['total_bill'])
propinas
```

```python
sns.lmplot('size', 'porcentaje', propinas)
```

```python
sns.lmplot('total_bill', 'porcentaje', propinas, hue = 'sex', markers = ['x', 'o'])
```

## 13.4 Mapas de Calor (Heatmaps)

https://seaborn.pydata.org/generated/seaborn.heatmap.html

Para construir un mapa de calor necesitamos una matriz de datos. 

```python
# Leer una base de datos de vuelos que está en nuestra librería Seaborn

vuelos  = sns.load_dataset('flights')
vuelos.head()
```

```python
vuelos.info()
```

```python
# Crearemos una matriz donde los renglones sean los meses y las columnas los años

vuelos = vuelos.pivot('month', 'year', 'passengers')
vuelos
```

```python
sns.heatmap(vuelos)
```

```python
sns.heatmap(vuelos, annot = True, fmt = 'd')
```

```python
valor_central = vuelos.loc['May'][1956]
valor_central
```

```python
sns.heatmap(vuelos, center = valor_central, annot = True, fmt = 'd')
```
