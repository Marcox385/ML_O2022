Guardar los datos como un dataframe y manipularlo de la siguiente forma:

1. Nombrar sus encabezados de la forma correcta
2. Borrar la última columna de notas
3. Borrar los registro que tengan datos nulos (NaN)

## Obtener los datos


```python
import pandas as pd
```

```python
url = 'https://es.wikipedia.org/wiki/Anexo:Finales_de_la_Copa_Mundial_de_F%C3%BAtbol'
```

```python
df = pd.io.html.read_html(url)
```

```python
df
```

```python
df_futbol = df[0]
```

```python
df_futbol
```

## 9.1. Nombrar sus encabezados de la forma correcta

```python
''' NOTA : Como se puede apreciar, el resultado para la celda anterior ya dio el resultado deseado.
    Sin embargo, a continuación se replicará el error esperado y se corregirá de igual manera '''

# Construir dataframe a corregir
headers = df_futbol.columns
df_futbol = df_futbol[:].values

holder = pd.DataFrame(None, columns = range(len(headers)))
holder.loc[0, :] = headers
for i, r in zip(range(len(df_futbol)), df_futbol):
    holder.loc[i + 1, :] = r
df_futbol = holder
df_futbol
```

```python
#Convertirlos en un diccionario
new_headers = df_futbol.loc[0, :].to_dict().values()
new_headers
```

```python
#Utilizar el diccionario para nombrar los encabezados
new_df_futbol = pd.DataFrame(df_futbol)
new_df_futbol.columns = new_headers
new_df_futbol
```

## 9.2. Borrar la última columna de notas

```python
new_df_futbol = new_df_futbol.iloc[:, :-1]
new_df_futbol
```

## 9.3. Borrar registros con datos nulos (NaN)



```python
new_df_futbol = new_df_futbol.dropna()
new_df_futbol
```
