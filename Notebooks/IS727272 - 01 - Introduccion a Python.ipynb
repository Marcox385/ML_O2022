## a) Identificadores 

Un identificador es el nombre que identifica tanto las variables como las funciones en un programa.

### Reglas:
- Es una secuencia de caracteres que consisten en letras, dígitos, y guiones bajos (_).
- Un identificador debe comenzar con una letra o un guión bajo (_), no puede comenzar con un número.
- Un identificador no puede ser una palabra clave o palabra reservada
- Puede ser de cualquier longitud.
- Debido a que Python es sensible a mayúsculas, area, Area y AREA son identificadores diferentes.


```python
numero = 5
cadena = "Hola Mundo"
Numero = 6

print(numero, Numero, cadena)
```

## b) Operadores

- Suma: +
- Resta: -
- Multiplicación: *
- División flotante: /
- Dvisión entera: //
- Exponenciación: **
- Residuo: %

```python
Suma = numero + Numero
Resta = 58 - Numero
Multiplicacion = numero * Numero
Division_flot = 58 / numero
Division_entera = 58 // 3
Exponienciacion = numero ** 3
Residuo = 58 % numero

print("Suma: ", Suma)
print("Resta: ", Resta)
print("Multiplicacion: ", Multiplicacion)
print("División flotante: ", Division_flot)
print("Division_entera: ", Division_entera)
print("Exponienciación: ", Exponienciacion)
print("Residuo: ", Residuo)
```

## d) Funciones matemáticas

Se utiliza el módulo math.

Para hacer uso de las funciones que están dentro de este módulo, librería o clase se debe mandar a llamar con la siguiente instrucción al inicio del programa en Python:

	import math
    
Y después se podrán utilizar las funciones con el siguiente formato:

	math.function(var)
    
También pueden utilizarse las constantes matemáticas pi y e como, que NO son funciones:

	math.pi
	math.e



```python
import math

print("FUNCIONES ALGEBRAICAS")

print("cos(PI / 2) = ", math.cos(math.pi / 2))
print("sin(PI / 2) = ", math.sin(math.pi / 2))
print("tan(PI / 2) = ", math.tan(math.pi / 2))

print("exp(1.0): ", math.exp(1))
print("log(e): ", math.log(math.e))
print("raíz de 4: ", math.sqrt(4))
print("Grados de PI/2: ", math.degrees(math.pi / 2))
print("Radianes de 90: ", math.radians(90))
```

La fórmula para calcular el área de un pentágono es: $Area = \frac{3\sqrt{3}}{2}s^2$, donde s es la longitud de un lado. El lado puede ser calculado utilizando la fórmula $s = 2*r* sin \frac{\pi}{5}$, donde r es la longitud del centro a un vértice del pentágono. 

```python
r = float(input("Ingrese el valor del radio: "))
print("El área del pentágono es {}".format(((3 * 3**(1/2))/2) * (2 * r * math.sin(math.pi / 5))**2))
```

## e) Cadenas y caracteres

Una cadena es una secuencia de caracteres y puede incluir texto y números.

Python trata a los caracteres y las cadenas de la misma manera.
Los valores de cadena deben estar encerrados en simples comillas (‘ ’) o comillas dobles (“ ”).

Ejemplos:

- Letter = ‘A’   #same as letter = “A”

- numChar = ‘4’  #same as numChar = “4”

- Message = “Good morning”   #same as message = ‘Good morning’

La función str convierte un número a una cadena:

	s = str(3.4)
	print(s)
	>>>’3.4’
	s2 = str(3)
	print(s2)
	>>> ‘3’

El operador de concatenación: +, se utiliza para unir o concatenar dos cadenas: 

    message = “Welcome” + “to” + “Python”
	print(message)
	>>> ‘Welcome to Python’

	chapterNo = 3
	s = “Chapter “ + str(chapterNo)
	print(s)
	>>> ‘Chapter 3’



## Ejercicio 2

Escribe un programa que lea la siguiente información:

    Nombre del empleado: 
    Número de horas trabajadas a la semana:
    Ingrese el pago por hora:
    Ingrese el porcentaje del impuesto federal:
    Ingrese el porcentaje del impuesto estatal:
    
E imprima el desglose de la nómina:

    Nombre del empleado: Jesús González
    Horas trabajadas: 10.0
    Pago por hora: $9.75
    Sueldo bruto: $97.5
    Deducciones:
        Impuesto federal (20.0%): $19.5
        Impuesto estatal (9.0%): $8.77
        Deducción total: $28.27
    Sueldo neto: $69.22

```python
name = input("Nombre del empleado: ")
hrs = float(input("Número de horas trabajadas a la semana: "))
pp_hr = float(input("Ingrese el pago por hora: "))
imp_fed = float(input("Ingrese el porcentaje del impuesto federal: "))
imp_est = float(input("Ingrese el porcentaje del impuesto estatal: "))

sueldo = hrs * pp_hr
imp_f_c = sueldo * (imp_fed/100)
imp_e_c = sueldo * (imp_est/100)
deduc = imp_f_c + imp_e_c

print("\nNombre del empleado: {}\n"
      "Horas trabajadas: {}\n"
      "Pago por hora: ${}\n"
      "Sueldo bruto: ${}\n"
      "Deducciones:\n"
      "\tImpuesto federal ({}%): ${}\n"
      "\tImpuesto estatal: ({}%): ${}\n"
      "\tDeducción total: ${}\n"
      "Sueldo neto: ${}".format(name, hrs, pp_hr, sueldo, imp_fed, imp_f_c, imp_est, imp_e_c, deduc, sueldo - deduc))
```

## f) Formateo de números y cadenas

- La función $format()$ se utiliza para regresar una cadena formateada.

- Sintaxis: format(item, format-specifier)

    - item es un número o una cadena
    - format-specifier es una cadena que especifica el formato.
    - La función devuelve un string.


```python
amount = 12618.98
interestRate = 0.0013
interest = amount * interestRate
print("Interest is", interest)
print("Interest is", round(interest, 2))
print("Interest is", format(interest, ".2f"))
```

```python
print(format(57.467657, "10.2f"))
print(format(12345678.923, "10.2f"))
print(format(57.4, "10.2f"))
print(format(57, "10.2f"))
```
